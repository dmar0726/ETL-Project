```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
exports = "Exports-calendar year.csv"
export_df = pd.read_csv(exports)
export_df
```

```python
export_df.dropna(subset=["Exports of all forms of coffee by all exporting countries"], how="all", inplace=True)
export_df
```

```python
export_df = export_df.drop(index=[0, 60, 61])
export_df.head()
```

```python
export_df["Exports of all forms of coffee by all exporting countries"] = export_df["Exports of all forms of coffee by all exporting countries"].replace({"Bolivia (Plurinational State of)": "Bolivia", "Timor-Leste": "East Timor", "Congo": "Congo, Repub. of the ", "Central African Republic": "Central African Rep", "Côte d'Ivoire": "Cote d'Ivoire", "Democratic Republic of Congo": "Congo, Dem. Rep.", "Lao People's Democratic Republic": "Laos", "Viet Nam": "Vietnam"})
export_df
```

```python
export_df.columns = export_df.iloc[0]
export_df
export_df.head()
```

```python
export_df = export_df.drop(2)
export_df.reset_index(drop=True)
export_df.head()
```

```python
export_df = export_df.rename(columns = {"Calendar years": "Country"})
export_df.head()
```

```python
final_df = export_df.set_index(keys="Country")
final_df.head()
```

```python
export_df.to_csv("exports_ico_clean.csv", header=True)
```
