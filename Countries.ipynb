```python
import pandas as pd
```

```python
PSD_Coffee_Data_file = "PSD Coffee Data.csv"
PSD_Coffee_Data_df = pd.read_csv(PSD_Coffee_Data_file)
PSD_Coffee_Data_df.head()
```

```python
# Create a filtered dataframe from specific columns
PSD_Coffee_Data_cols = ["Country_Name", "Market_Year", "Calendar_Year", "Month", "Attribute_ID", "Attribute_Description", "Unit_ID", "Unit_Description", "Value"]
PSD_Coffee_Data_transformed= PSD_Coffee_Data_df[PSD_Coffee_Data_cols].copy()

# Rename the column headers
PSD_Coffee_Data_transformed = PSD_Coffee_Data_transformed.rename(columns={"Country_Name": "Country"})

PSD_Coffee_Data_transformed.index
```
