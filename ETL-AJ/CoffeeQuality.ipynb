### Clean DataFrame

```python
# Rename column names for arabica_df2 and robusta_df2.
robusta_df3 = robusta_df2.rename(columns = {"Fragrance / Aroma": "Aroma",
                             "Salt / Acid": "Acidity",
                             "Uniform Cup": "Uniformity",
                             "Country of Origin": "Country",
                             "Bitter / Sweet": "Sweetness",
                             "quality_score": "Quality Score",
                             "Mouthfeel": "Body",
                             "Expiration": "Expiration of Certification"}) 

arabica_df3 = arabica_df2.rename(columns = {"quality_score": "Quality Score",
                                            "Country of Origin": "Country",
                                           "Expiration": "Expiration of Certification"}) 
                                  
```

```python
#Total rows in both robusta_df3 and arabica_df3.

total_rows = len(robusta_df3) + len (arabica_df3)

total_rows
```

```python
#Join arabica_df3 with robusta_df3.  
merged_df = pd.concat([arabica_df3, robusta_df3], ignore_index=True)
merged_df
```

```python
#Filter out any data which is the Status is "Incomplete".
print(merged_df['Status'].unique())

#All data are complete.  
```

```python
#Delete Status column from merged_df.  
list(merged_df.columns)

merged_df2 = merged_df[['Species','Variety','Farm Name','Country','Region','Altitude','Owner','Company','Mill','Producer','In-Country Partner','Harvest Year','Processing Method','Aroma','Flavor','Aftertaste','Acidity','Body','Balance','Uniformity','Clean Cup','Sweetness','Cupper Points','Total Cup Points','Quality Score','Moisture','Category One Defects','Quakers','Color','Category Two Defects','Expiration of Certification']]

merged_df2
```

```python
merged_df2.dtypes
```

```python
#Convert the following columns from object to str to format the datapoints (see below).

merged_df2.astype({
    'Species': str,
    'Altitude': str,
    'Harvest Year': str,
    'Total Cup Points': str,
    'Moisture': str,
    'Category One Defects': str,
    'Category Two Defects': str,
}).dtypes
```

```python
# Column Species: Delete extra "b''" from Arabica and Rubusta.
merged_df2['Species'].unique()
merged_df2['Species'].replace(["b'Arabica'", "b'Robusta'"], ["Arabica", "Robusta"])
```

```python
merged_df2.to_csv("Coffee_Quality_v1.csv")

# Species Column is formatted correctly.
```

```python
# Total Cup Points Column: Remove Sample from all entries.
merged_df2['Total Cup Points'] = merged_df2['Total Cup Points'].map(lambda x: x.lstrip('Sample'))
merged_df2['Total Cup Points'].astype(float)
```

```python
# Moisture Column: Remove % from all entries.
merged_df2['Moisture'] = merged_df2['Moisture'].map(lambda x: x.rstrip('%'))
merged_df2['Moisture'].astype(float)
```

```python
# Category One Defects Column: Remove % from all entries.
merged_df2['Category One Defects'] = merged_df2['Category One Defects'].map(lambda x: x.rstrip('full defects'))
merged_df2['Category One Defects'].astype(int)
```

```python
# Category Two Defects Column: Remove % from all entries.
merged_df2['Category Two Defects'] = merged_df2['Category Two Defects'].map(lambda x: x.rstrip('full defects'))
merged_df2['Category Two Defects'].astype(int)
```

```python
merged_df2.dtypes
```

```python
merged_df2.to_csv("Coffee_Quality_v2.csv")

# Total Cup Points, Moisture, Category One, and Category Two Columns are formatted correctly.
```

```python
merged_df2["Harvest Year"].unique()
```

```python
#Delete month and date from Harvest Year.
merged_df2["Harvest Year"] = merged_df2["Harvest Year"].replace(['Fall 2009'], '2009')
merged_df2["Harvest Year"] = merged_df2["Harvest Year"].replace(['March 2010','4T/10', '23 July 2010', '4t/2010', '4T/2010', '47/2010', '4T72010'], '2010')
merged_df2["Harvest Year"] = merged_df2["Harvest Year"].replace(['January 2011', '1T/2011','1t/2011','4t/2011', 'Abril - Julio /2011', 'Spring 2011 in Colombia.', '3T/2011'], '2011')
merged_df2["Harvest Year"] = merged_df2["Harvest Year"].replace(['Sept 2009 - April 2010', 'December 2009-March 2010'], '2009/2010')
merged_df2["Harvest Year"] = merged_df2["Harvest Year"].replace(['08/09 crop'], '2008/2009')
merged_df2["Harvest Year"].unique()
```

```python
#Delete "non-sense"/incomplete data from Harvest Year.
merged_df2.drop(merged_df2.index[merged_df2["Harvest Year"] == 'mmm'], inplace = True)
merged_df2.drop(merged_df2.index[merged_df2["Harvest Year"] == 'May-August'], inplace = True)
merged_df2.drop(merged_df2.index[merged_df2["Harvest Year"] == 'TEST'], inplace = True)
merged_df2.drop(merged_df2.index[merged_df2["Harvest Year"] == 'January Through April'], inplace = True)
merged_df2.drop(merged_df2.index[merged_df2["Harvest Year"] == 'August to December'], inplace = True)
merged_df2.drop(merged_df2.index[merged_df2["Harvest Year"] == 'Mayo a Julio'], inplace = True)
merged_df2.drop(merged_df2.index[merged_df2["Harvest Year"] == 'Abril - Julio'], inplace = True)
merged_df2["Harvest Year"].unique()
```

```python
#Covert all '-' into '/' and have same formatting in Harvest Year.
merged_df2["Harvest Year"] = merged_df2["Harvest Year"].replace(['2009 / 2010', '2009 - 2010', '2009-2010'], '2009/2010')
merged_df2["Harvest Year"] = merged_df2["Harvest Year"].replace(['2010-2011'], '2010/2011')
merged_df2["Harvest Year"] = merged_df2["Harvest Year"].replace(['2016 / 2017'], '2016/2017')
merged_df2["Harvest Year"] = merged_df2["Harvest Year"].replace(['2017 / 2018'], '2017/2018')
merged_df2["Harvest Year"].unique()
```

```python
merged_df2.to_csv("Coffee_Quality_v3.csv")

# Harvest Year column is formatted correctly.
```

```python
merged_df2["Altitude"].unique()
```

```python
#Delete all variants of meter labels from Altitude.
merged_df2["Altitude"] = merged_df2["Altitude"].str.replace('[ppx, B, +, ~, g, M, MSNM, msm, mosl, METROS, mals, MSN, SNN, Meters, m, meters, mts, msnm, 公尺, M公尺, MTRS. A NIVEL DEL MAR, Meters above sea level, :]', '')

merged_df2["Altitude"].unique() 
```

```python
# Convert feet to meters in Altitude column.
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['2000f'], '610')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['1400f'], '427')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['3500-5200f'], '1067-1585')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['4500-6000F'], '1372-1829')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['5500-6000P'], '1676-1829')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['1200-1300dp'], '366-396')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['4500Pi'], '1372')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['5000Pi', '5000F'], '1524')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['1400-1900P'], '427-579')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['4500Pi'], '1372')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['4400-4700f'], '1341-1433')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['3600-6200f'], '1097-1890')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['4000f', '4000P'], '1219')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['3500f'], '1067')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['2800f'], '853')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['4921f'], '1500')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['4200f', '4200F'], '1280')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['4600f'], '1402')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['5600f'], '1707')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['3000f'], '914')
merged_df2["Altitude"].unique()
```

```python
# Convert range data to same format in Altitude column.
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['13001400'], '1300-1400')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['16001950'], '1600-1950')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['16001900'], '1600-1900')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['14501700'], '1450-1700')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['d9001100'], '900-1100')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['ppx1500'], '500')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['13001800'], '1300-1800')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['1400hu1850'], '1400-1850')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['1800(5900'], '1800')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['12001400'], '1200-1400')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['13001500'], '1300-1500')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['w700d1400'], '700-1400')
merged_df2["Altitude"] = merged_df2["Altitude"].replace(['12001350'], '1200-1350')

merged_df2["Altitude"].unique()
```

```python
#Delete "non-sense"/incomplete data from Altitude.
merged_df2.drop(merged_df2.index[merged_df2["Altitude"] == '-1'], inplace = True)
merged_df2.drop(merged_df2.index[merged_df2["Altitude"] == 'ukw'], inplace = True)
merged_df2.drop(merged_df2.index[merged_df2["Altitude"] == '-(g)'], inplace = True)
merged_df2.drop(merged_df2.index[merged_df2["Altitude"] == 'Huuc'], inplace = True)
merged_df2.drop(merged_df2.index[merged_df2["Altitude"] == 'ppx1'], inplace = True)
merged_df2.drop(merged_df2.index[merged_df2["Altitude"] == '-'], inplace = True)
merged_df2["Altitude"].unique()
```

```python
merged_df2 = merged_df2.rename(columns = {"Altitude": "Altitude (Meters)"}) 
```
