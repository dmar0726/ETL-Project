```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
Countries = pd.read_csv("country_codes.csv")
Coffee_Data_file = pd.read_csv("Coffee_Data_df_FINAL.csv")
Coffee_Quality_FINAL = pd.read_csv("Coffee_Quality_FINAL.csv")
domestic_consumption_ico_clean = pd.read_csv("domestic_consumption_ico_clean.csv")
exports_ico_clean = pd.read_csv("exports_ico_clean.csv")
import_consumption_ico = pd.read_csv("import_consumption_ico.csv")
imports_ico = pd.read_csv("imports_ico.csv")
inventories_ico = pd.read_csv("inventories_ico.csv")
```

```python
connection_string = f'postgresql://postgres:Syracuse10!@localhost:5432/Coffee_db'
engine = create_engine(connection_string)
```

```python
engine.table_names()
```

```python
Countries.to_sql(name='Countries', con=engine, if_exists='replace', index=False)
```

```python
pd.read_sql_query('select * from "Countries"', con=engine).head()
```

```python
Coffee_Data_file.to_sql(name='US_Coffee_Imports', con=engine, if_exists='replace', index=False)
```

```python
pd.read_sql_query('select * from "US_Coffee_Imports"', con=engine).head()
```

```python
Coffee_Quality_FINAL.to_sql(name='Coffee_Farms_Quality', con=engine, if_exists='replace', index=False)
```

```python
pd.read_sql_query('select * from "Coffee_Farms_Quality"', con=engine).head()
```
