```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
domestic_consumption = "Domestic-consumption.csv"
domestic_consumption_df = pd.read_csv(domestic_consumption)
domestic_consumption_df
```

```python
domestic_consumption_df.dropna(subset=["Domestic consumption by all exporting countries"], how="all", inplace=True)
domestic_consumption_df
```

```python
new_dc_df = domestic_consumption_df.drop(index=[0, 4, 20, 29, 66, 67])
new_dc_df
```

```python
new_dc_df.columns = new_dc_df.iloc[0]
```

```python
new_dc_df.head()
```

```python
new_dc_df = new_dc_df.drop(2)
new_dc_df.reset_index(drop=True)
```

```python
new_dc_df = new_dc_df.rename(columns = {"Crop year": "Producer"})
new_dc_df.head()
```

```python
new_dc_df.set_index(keys="Producer")
new_dc_df
```

```python
final_df = new_dc_df.drop(new_dc_df.columns[[1]], axis=1)
final_df
```

```python
final_df.set_index(keys="Producer")
```

```python
final_df.to_csv("domestic_consumption_ico_clean.csv", header=True)
```
